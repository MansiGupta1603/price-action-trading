```python
data['local_max'] =data['Close1'][(data['Close1'].shift(1)<data['Close1'])&(data['Close1'].shift(-1)<data['Close1'])]
data['local_min'] =data['Close1'][(data['Close1'].shift(1)>data['Close1'])&(data['Close1'].shift(-1)>data['Close1'])]
colors = plt.rcParams['axes.prop_cycle'].by_key()['color']
plt.figure(figsize=(15, 8))
plt.plot(data['Close1'], zorder=0)
plt.scatter(data.index, data['local_max'], s=100,
  label='Maxima', marker='^', c=colors[1])
plt.scatter(data.index, data['local_min'], s=100,
  label='Minima', marker='v', c=colors[2])
plt.xlabel('Date')
plt.ylabel('Price ($)')

plt.legend()
plt.show()
```

```python
from numpy import array
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import Dense,Dropout


def split_sequence(sequence, n_steps_in, n_steps_out):
	X, y = list(), list()
	for i in range(len(sequence)-10000):
		
		end_ix = i + n_steps_in
		out_end_ix = end_ix + n_steps_out
		
		if out_end_ix > len(sequence):
			break
	
		seq_x, seq_y = sequence[i:end_ix], sequence[end_ix:out_end_ix]
		X.append(seq_x)
		y.append(seq_y)
	return array(X), array(y)



n_steps_in, n_steps_out = 50, 2

X, y = split_sequence(cp['returns'], n_steps_in, n_steps_out)

n_features = 1
X = X.reshape((X.shape[0], X.shape[1], n_features))

model = Sequential()
model.add(LSTM(100, activation='relu', return_sequences=True, input_shape=(n_steps_in, n_features)))
model.add(Dropout(0.1))
model.add(LSTM(100, activation='relu'))
model.add(Dropout(0.1))
model.add(Dense(n_steps_out))
model.compile(optimizer='adam', loss='mse')
# fit model
model.fit(X, y, epochs=20, verbose=1)

```

```python
def split_sequence2(sequence, n_steps_in, n_steps_out):
	X, y = list(), list()
	for i in range(len(sequence)-5000,len(sequence)):
		
		end_ix = i + n_steps_in
		out_end_ix = end_ix + n_steps_out
		
		if out_end_ix > len(sequence):
			break
	
		seq_x, seq_y = sequence[i:end_ix], sequence[end_ix:out_end_ix]
		X.append(seq_x)
		y.append(seq_y)
	return array(X), array(y)



n_steps_in, n_steps_out = 50, 2

X1, y1 = split_sequence2(cp['returns'], n_steps_in, n_steps_out)

n_features = 1
X1 = X1.reshape((X1.shape[0], X1.shape[1], n_features))
```
