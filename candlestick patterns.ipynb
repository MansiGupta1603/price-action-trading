```
def profits2(df,price,signal):
  bbBuy = []
  bbSell = []
  bbBuy2 = []
  bbSell2 = []
  position = False
  for i in range(1, len(price)):
    if signal[i] ==1:
      if position == False :
       bbBuy.append(df['Close'][i])
       bbBuy2.append(i)
       position = True
    elif signal[i] == -1:
      if position == True :
       bbSell.append(df['Close'][i])
       bbSell2.append(i)
       position = False
  n = min(len(bbBuy),len(bbSell))
  profitsrel = []
  for i in range(0,n):
    profits = (bbSell[i]-bbBuy[i])*100/(bbBuy[i]+bbSell[i])
    profitsrel.append(profits)
  return profitsrel,bbBuy2,bbSell2



                
              
```
