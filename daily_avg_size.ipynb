```python
data['body1']= (data['Close']- data['Open'])/(data['Close'])
data['body']= abs(data['Close']- data['Open'])/(data['Close'])
```

```python
import numpy as np
from scipy.stats import norm

# Plot a normal distribution of the 'body' values
mu, std = norm.fit(data['body1'])
plt.hist(data['body1'], bins=25, density=True, alpha=0.6, color='g')
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'k', linewidth=2)
title = "Fit results: mu = %.2f,  std = %.2f" % (mu, std)
plt.title(title)
plt.xlabel("Body")
plt.ylabel("Frequency")
plt.show()
```
